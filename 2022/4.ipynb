```python
input = open("inputs/4").read()
```

```python
def make_range(s):
    a, b = map(int, s.split("-"))
    return set(range(a, b+1))
```

```python
# pairs = [tuple(map(make_range, s.split(','))) for s in test_input.split()]
pairs = [tuple(map(make_range, s.split(','))) for s in input.split()]
```

```python
s = 0
for s1, s2 in pairs:
    i = s1 & s2

    if (i == s1) or (i == s2):
        s += 1

s
```
