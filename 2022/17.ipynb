```python
import itertools
```

```python
test_input = """####

.#.
###
.#.

..#
..#
###

#
#
#
#

##
##"""
```

```python
WIDTH = 7
FLOOR = set((i, 0) for i in range(7))
X_OFFSET = 2
Y_OFFSET = 3
```

```python
rock_sets = []


for pattern in test_input.split("\n\n"):
    rock_set = set()

    levels = pattern.split('\n')
    num_levels = len(levels)

    for i, level in enumerate(levels):
        for j, c in enumerate(level):
            if c == "#":
                rock_set.add((j+X_OFFSET, num_levels - i - 1))
    
    rock_sets.append(rock_set)

num_rock_sets = len(rock_sets)
```

```python
# jet_pattern = ">>><<><>><<<>><>>><<<>>><<<><<<>><>><<>>"
jet_pattern = open("inputs/17").read()
```

```python
rock_sets
```

```python
current_top = 0
resting_rocks = set()

jet_iterator = itertools.cycle(enumerate(jet_pattern))

top_dict = {}

for i in itertools.count(0):
    rock_cycle = i % num_rock_sets
    rock = rock_sets[rock_cycle]

    # X_offset already baked in above; put that here? hm
    falling_rock = set((x, current_top+Y_OFFSET+y) for (x, y) in rock)
    # print(f"Rock: {rock}")
    # print(f"Falling rock: {falling_rock}")

    while True:
        # gets pushed by jet
        jet_cycle, jet = next(jet_iterator)
        direction = 1 if jet == ">" else -1
        
        pushed_rock = set((x+direction, y) for (x, y) in falling_rock)

        # check if pushing left or right would intersect with resting rocks or the wall
        if any(x < 0 or x > WIDTH-1 for (x, y) in pushed_rock) or pushed_rock & resting_rocks:
            pass
        else:
            falling_rock = pushed_rock
        
        # now try falling one
        fallen_one = set((x, y-1) for (x, y) in falling_rock)

        # make sure not intersected floor or resting rock
        if any(y < 0 for (x, y) in fallen_one) or fallen_one & resting_rocks:
            # at rest; store last position and break
            # WHAT ABOUT UNDOING THE JET TIMER? UH
            resting_rocks |= falling_rock
            break
        else:
            falling_rock = fallen_one
    
    current_top = max(y+1 for (x, y) in resting_rocks)
    # print(f"After {i} fell the current top is {current_top}")
    # print(f"Resting rocks: {resting_rocks}")
    # print()

    top_dict[i] = current_top

    new_floor = all((i, current_top-1) in resting_rocks for i in range(WIDTH))

    if new_floor:
        print(f"New floor at iteration {i}. Current top: {current_top}. Rock cycle: {rock_cycle}. Jet cycle: {jet_cycle}")
    
    # print(f"Iteration {i}. Current top: {current_top}. Rock cycle: {rock_cycle}. Jet cycle: {jet_cycle}")

    if i+1 == 10_000:
        break
```
