```python
import numpy as np

test_input = """R 4
U 4
L 3
D 1
R 4
D 1
L 5
R 2"""

test_input2 = """R 5
U 8
L 8
D 3
R 17
D 10
L 25
U 20"""

input = open("inputs/9").read()
```

```python
dirs = []

# for s in test_input.split("\n"):
# for s in test_input2.split("\n"):
for s in input.split("\n"):
    a, b = s.split()

    dirs.append((a, int(b)))
```

```python
def print_board(positions):
    positions = np.array(positions)

    min_x, min_y = positions.min(axis=0)
    positions -= np.array([min_x, min_y])

    max_x, max_y = positions.max(axis=0)

    board = np.full((max_y+1, max_x+1), ".")

    for n, (i, j) in enumerate(positions[::-1]):
        board[j, i] = str(len(positions)-n-1) if n != len(positions)-1 else 'H'

    print(board)
```

```python
def p1():
    head_position = np.zeros(2, dtype=np.int32)
    tail_position = np.zeros(2, dtype=np.int32)

    all_tail_positions = set([tuple(tail_position)])

    for dir, dist in dirs:
        # move head
        if dir == 'U':
            move_dir = np.array([0, 1])
        elif dir == 'D':
            move_dir = np.array([0, -1])
        elif dir == 'L':
            move_dir = np.array([-1, 0])
        else:
            move_dir = np.array([1, 0])
        
        # honestly I shouldn't need to loop thru each piece of the dist one by one
        for d in range(dist):
            head_position += move_dir
            diff = head_position - tail_position
            abs_diff = np.abs(diff)

            if any(abs_diff > 1):
                tail_position = head_position - move_dir

                all_tail_positions.add(tuple(tail_position))
            
    return len(all_tail_positions)


def p1_2():
    head_position = np.zeros(2, dtype=np.int32)
    tail_position = np.zeros(2, dtype=np.int32)

    all_tail_positions = set([tuple(tail_position)])

    for dir, dist in dirs:
        # move head
        if dir == 'U':
            move_dir = np.array([0, 1])
        elif dir == 'D':
            move_dir = np.array([0, -1])
        elif dir == 'L':
            move_dir = np.array([-1, 0])
        else:
            move_dir = np.array([1, 0])
        
        # honestly I shouldn't need to loop thru each piece of the dist one by one
        for d in range(dist):
            head_position += move_dir
            diff = head_position - tail_position
            abs_diff = np.abs(diff)

            if any(abs_diff > 1):
                tail_position += np.nan_to_num(diff // abs_diff)

                all_tail_positions.add(tuple(tail_position))
            
    return len(all_tail_positions)
```
