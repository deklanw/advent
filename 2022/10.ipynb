```python
def clock_generator(instructions):
    X = 1
    cycle = 1

    for s in instructions:
        yield cycle, X

        if s == "noop":
            cycle += 1
        else:
            summand = int(s.split(" ")[1])

            cycle += 1
            yield cycle, X
            cycle += 1
            X += summand

    yield cycle, X
```

```python
# instructions = test_input.split('\n')
# instructions = test_input2.split('\n')
instructions = open("inputs/10").read().split('\n')

sum(cycle * x for cycle, x in clock_generator(instructions) if (cycle % 40 == 20) and cycle <= 220)
```

```python
def print_screen(a):
    screen = np.reshape(a, (6, 40))
    m, n = screen.shape

    for i in range(m):
        row_s = "".join('#' if screen[i, j] > 0 else '.' for j in range(n))
        print(row_s)
```

```python
screen = np.zeros(6 * 40)

for (cycle, X) in clock_generator(instructions):
    sprite_position = X
    draw_position = cycle - 1

    if abs((draw_position % 40) - sprite_position) <= 1:
        screen[draw_position] = 1

print_screen(screen)
```
