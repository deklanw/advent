```python
import itertools
import networkx as nx
```

```python
def parse_input(s):
    return [tuple(map(int, ss.split(","))) for ss in s.split("\n")]
```

```python
def neighbors(t):
    x, y, z = t

    return set([(x+1, y, z),
        (x-1, y, z),
        (x, y+1, z),
        (x, y-1, z),
        (x, y, z+1),
        (x, y, z-1)])
```

```python
def p1(cubes):
    num_exposed_faces = 0
    isolated_points = set()

    def get_candidate_neighboring(i):
        # everything but itself works, logically
        # could speed this up
        return cubes[:i] + cubes[i+1:]

    for i, c in enumerate(cubes):
        # print(c)
        # get candidate neighboring cubes
        candidates = get_candidate_neighboring(i)

        neighboring = set(candidates)
        # print(neighboring)
        e = neighbors(c) - neighboring

        if len(e) == 6:
            isolated_points.add(c)

        # print(e)
        num_exposed_faces += len(e)
        # print()

    return num_exposed_faces
```

```python
p1(cubes)
```

```python
def p2(cubes):
    extrema = []

    for j in range(3):
        p = [t[j] for t in cubes]
        extrema.append((min(p), max(p)))

    full_space = set(itertools.product(*[range(x-1, y+2) for (x, y) in extrema]))
    cube_set = set(cubes)
    air = full_space - cube_set


    air_g = nx.Graph()

    for a in air:
        for n in neighbors(a):
            if n in air:
                air_g.add_edge(a, n)


    all_components = list(nx.connected_components(air_g))
    sample_air_cube = tuple(m-1 for (m, n) in extrema)
    outer_air = next(c for c in all_components if sample_air_cube in c)

    solid_mass = full_space - outer_air

    return p1(list(solid_mass))
```
