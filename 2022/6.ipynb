```python
from collections import deque
```

```python
input = open("inputs/6").read()
```

```python
def p(n):
    q = deque(input[:n])

    for i, c in enumerate(input[n:]):
        if len(set(q)) == n:
            return i+n

        q.popleft()
        q.append(c)

def p2(n):
    for i in range(len(input)-n):
        if len(set(input[i:(i+n)])) == n:
            return i+n
```

```python
p(4), p(14)
```

```python
p2(4), p2(14)
```
