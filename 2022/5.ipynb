```python
import re
```

```python
input = open("inputs/5").read()
```

```python
instruction_re = r"move\s(\d+)\sfrom\s(\d+)\sto\s(\d+)"
```

```python
# hardcoding this, don't care
stacks = [
    'RPCDBG',
    'HVG',
    'NSQDJPM',
    'PSLGDCNM',
    'JBNCPFLS',
    'QBDZVGTS',
    'BZMHFTQ',
    'CMDBF',
    'FCQG'
]
stacks = [list(s) for s in stacks]
```

```python
instructions = []

for l in input.split('\n'):
    m = re.match(instruction_re, l)

    if m:
        instructions.append(tuple(map(int, m.groups())))
```

```python
def puzzle(stacks, instructions, ord=-1):
    moved_stacks = [st.copy() for st in stacks]

    for (amount, start, end) in instructions:
        to_move = moved_stacks[start-1][-amount:][::ord]
        moved_stacks[end-1].extend(to_move)
        del moved_stacks[start-1][-amount:]

    return "".join(l[-1] for l in moved_stacks)
```

```python
# p1
puzzle(stacks, instructions)
```

```python
# p2
puzzle(stacks, instructions, ord=1)
```
