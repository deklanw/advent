```python
def parse(input_s):
    return [(ast.literal_eval(a), ast.literal_eval(b)) for (a, b) in map(lambda s: s.split("\n"), input_s.split("\n\n"))]

def parse_no_spaces(input_s):
    input_s = input_s.replace("\n\n", "\n")

    return [ast.literal_eval(s) for s in input_s.split("\n")]
```

```python
def compare(a, b):
    if type(a) == int and type(b) == int:
        if a < b:
            return -1
        elif a == b:
            return 0
        else:
            return 1
    
    if type(a) == list and type(b) == list:
        for i in itertools.count(0):
            # ran out simultaneously: inconclusive
            if (i == len(a)) and (i == len(b)):
                return 0

            if i > len(a) - 1:
                # a ran out
                return -1

            if i > len(b) - 1:
                # b ran out
                return 1

            c = compare(a[i], b[i])

            # if not tied, return result
            # if equal, inconclusive: keep looping
            if c != 0:
                return c

    elif type(a) == list:
        return compare(a, [b])
    else:
        return compare([a], b)
```

```python
sm = []

for i, (a, b) in enumerate(to_compare):
    c = compare(a, b)

    if c == -1:
        sm.append(i+1)
```
