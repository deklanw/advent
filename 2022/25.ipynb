```python
import itertools
```

```python
BASE = 5

summ = 0

# for l in test_input.split("\n"):
for l in open("inputs/25").read().split("\n"):
    s = 0

    for i, d in enumerate(l[::-1]):
        if d == '-':
            v = -1
        elif d == '=':
            v = -2
        else:
            v = int(d)
        
        s += BASE**i * v
    
    print(s)

    summ += s

summ
```

```python
def euclid(a, b):
    q = a // b

    return q, a - q*b
```

```python
start = summ
```

```python
# regular number system, for comparison
start_i = next(i for i in itertools.count() if BASE**i <= start and BASE**(i+1) > start)

digits = []

remainder = start

for i in range(start_i, -1, -1):
    digit, remainder = euclid(remainder, BASE**i)

    digits.append(digit)
```

```python
start_i = next(i for i in itertools.count() if 2*BASE**(i-1) < start and 2*BASE**i >= start)

digits = []

remainder = start

for i in range(start_i, -1, -1):
    print(remainder)

    digit = min((abs(remainder - c*BASE**i), c) for c in range(2, -3, -1))[1]

    print(i, remainder, digit, BASE**i)

    remainder -= digit * BASE**i

    digits.append(digit)
```

```python
def to_str(d):
    if d == -2:
        return "="
    elif d == -1:
        return "-"
    else:
        return str(d)
```
