```python
input = open("inputs/2").read()
```

```python
R = "R"
P = "P"
S = "S"
```

```python
points_map = {R: 1, P: 2, S: 3}
str_map = {"A": R, "B": P, "C": S, "X": R, "Y": P, "Z": S}

# used for p2 only
result_map = {R: 0, P: 3, S: 6}
```

```python
# Rock defeats Scissors, Scissors defeats Paper, and Paper defeats Rock.
cases = {
    (R, S): 0,
    (S, P): 0,
    (P, R): 0,
    (P, P): 3,
    (R, R): 3,
    (S, S): 3,
    (S, R): 6,
    (P, S): 6,
    (R, P): 6
}
```

```python
def p1(a, b):
    return cases[a,b] + points_map[b]

def p2(a, b):
    desired = result_map[b]

    for sh in points_map:    
        if cases[a, sh] == desired:
            return desired + points_map[sh]
```

```python
games = [[str_map[ss] for ss in s.split()] for s in input.split('\n')]
```

```python
sum([p1(*g) for g in games])
```

```python
sum([p2(*g) for g in games])
```
