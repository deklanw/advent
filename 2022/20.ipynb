```python
from collections_extended import bijection

def parse(s):
    return list(map(int, s.split("\n")))
```

```python
def mix(lst, rounds):
    N = len(lst)
    mapping = bijection({idx: idx for idx in range(N)})

    for round in range(rounds):
        for og_idx, n in enumerate(lst):
            # get current index of number
            start_idx = mapping[og_idx]

            shift = n % (N-1)

            # shift such that the relevant og_idx is now at current_idx 0
            update = {}
            for o, c in mapping.items():
                update[o] = (c-start_idx) % N
            mapping.update(update)

            update = {og_idx: shift}
            # move everything which is between appropriately
            for c in range(1, shift+1):
                update[mapping.inverse[c]] = c-1
            mapping.update(update)

    reordered = [lst[v] for (idx, v) in sorted([(idx, v) for v, idx in mapping.items()])]

    z = reordered.index(0)
    s = 0

    for g in [1000, 2000, 3000]:
        a = reordered[(z+g) % N]
        print(a)
        s += a

    return s
```
