```python
import numpy as np
import itertools
```

```python
test_input = """498,4 -> 498,6 -> 496,6
503,4 -> 502,4 -> 502,9 -> 494,9"""
```

```python
def get_paths(input):
    return [list(map(lambda ss: tuple(map(int, ss.split(','))), s.split(" -> "))) for s in input.split("\n")]
```

```python
source = (500,0)
```

```python
DOWN = np.array([0, 1])
DOWN_LEFT = np.array([-1, 1])
DOWN_RIGHT = np.array([1, 1])
```

```python
def min_max(a, b):
    return min(a, b), max(a, b)
```

```python
def p1(paths):
    sand = set()
    rock = set()

    # clean this mess up
    for p in paths:
        for (a, b), (c, d) in zip(p, p[1:]):
            if a == c:
                m, n = min_max(b, d)
                for i in range(m, n+1):
                    rock.add((a, i))
            else:
                m, n = min_max(a, c)
                for i in range(m, n+1):
                    rock.add((i, b))

    deepest = max(t[1] for t in rock)

    blocked = rock | sand

    for sand_num in itertools.count(0):
        sand_position = np.array(source)
        falling_forever = False

        while True:
            if tuple(sand_position + DOWN) not in blocked:
                sand_position += DOWN
            elif tuple(sand_position + DOWN_LEFT) not in blocked:
                sand_position += DOWN_LEFT
            elif tuple(sand_position + DOWN_RIGHT) not in blocked:
                sand_position += DOWN_RIGHT
            else:
                # print("Nowhere for sand to fall. At rest.")
                break
            
            if sand_position[1] > deepest:
                # print("Past the lowest rock. Falling forever. Breaking.")
                falling_forever = True
                break

        if falling_forever:
            # print("Falling forever. Breaking out of outer loop")
            break
        else:
            # The sand came to rest, note its position
            blocked.add(tuple(sand_position))
    
    return sand_num
```

```python
p1(get_paths(test_input))
```
