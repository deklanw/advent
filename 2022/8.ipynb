```python
N = grid.shape[0]

def vis(a):
    return np.array([a[i] > (a[:i].max(axis=0) if i > 0 else -1) for i in range(N)])
```

```python
top = vis(grid)
left = vis(grid.T).T

bottom = vis(grid[::-1])[::-1]
right = vis(grid[:, ::-1].T).T[:, ::-1]
```

```python
# 1792
(top | left | bottom | right).sum()
```

```python
def index_first_true(a):
    return next((i+1 for i, v in enumerate(a) if v), len(a))
```

```python
right_space = np.zeros_like(grid)
left_space = np.zeros_like(grid)
above_space = np.zeros_like(grid)
below_space = np.zeros_like(grid)

# just do right direction for simplicity
for (i, j), v in np.ndenumerate(grid):
    left_space[i, j] = index_first_true(grid[i, :j][::-1] >= v)
    right_space[i, j] = index_first_true(grid[i, (j+1):] >= v)

    above_space[i, j] = index_first_true(grid[:i, j][::-1] >= v)
    below_space[i, j] = index_first_true(grid[(i+1):, j] >= v)
```

```python
# 334880
(left_space * right_space * above_space * below_space).max()
```
