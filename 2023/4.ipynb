```python
test_input = """Card 1: 41 48 83 86 17 | 83 86  6 31 17  9 48 53
Card 2: 13 32 20 16 61 | 61 30 68 82 17 32 24 19
Card 3:  1 21 53 59 44 | 69 82 63 72 16 21 14  1
Card 4: 41 92 73 84 69 | 59 84 76 51 58  5 54 83
Card 5: 87 83 26 28 32 | 88 30 70 12 93 22 82 36
Card 6: 31 18 13 56 72 | 74 77 10 23 35 67 36 11"""
```

```python
def parse_card(card):
    winning_numbers, have_numbers = card.split(": ")[-1].split(" | ")

    winning_numbers = [int(x) for x in winning_numbers.split()]
    have_numbers = [int(x) for x in have_numbers.split()]

    # just checking!
    assert len(winning_numbers) == len(set(winning_numbers))
    assert len(have_numbers) == len(set(have_numbers))

    return set(winning_numbers), set(have_numbers)
```

```python
def pt1(input):
    cards = [parse_card(card) for card in input.splitlines()]

    s = 0
    for winning_numbers, have_numbers in cards:
        num_intersection = len(winning_numbers & have_numbers)

        if num_intersection > 0:
            s += 2**(num_intersection-1)

    return s
```

```python
pt1(test_input)
```

```python
pt1(input)
```

```python
from functools import lru_cache

def pt2(input):
    cards = [parse_card(card) for card in input.splitlines()]
    num_intersections = [len(winning_numbers & have_numbers) for winning_numbers, have_numbers in cards]

    @lru_cache(maxsize=None)
    def card_value(index):
        if index > len(num_intersections) - 1:
            return 0

        return 1 + sum(card_value(index+k+1) for k in range(num_intersections[index]))
    
    r = sum(card_value(i) for i in range(len(num_intersections)))

    print(card_value.cache_info())

    return r
```

```python
pt2(test_input)
```

```python
pt2(input)
```
