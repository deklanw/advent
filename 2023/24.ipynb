```python
import itertools
import numpy as np

s = 0

for (ra, va), (rb, vb) in itertools.combinations(input_rays, 2):
    ra = np.array(ra[:2])
    va = np.array(va[:2])
    rb = np.array(rb[:2])
    vb = np.array(vb[:2])

    same_direction = all(a / b == va[0] / vb[0] for a, b in zip(va, vb))

    if same_direction:
        print("Same direction")
        continue

    A = np.array([va, -vb]).T
    b = rb - ra

    t_a, t_b = np.linalg.solve(A, b)

    if t_a < 0 or t_b < 0:
        print("Backwards in time")
        continue

    x_h = ra[0] + t_a * va[0]
    y_h = ra[1] + t_a * va[1]

    if x_h < MIN_COORD or x_h > MAX_COORD or y_h < MIN_COORD or y_h > MAX_COORD:
        print("Out of bounds")
        continue

    s += 1

s
```

```python
from z3 import Real, Solver, And, sat, Int, Not, set_param
import numpy as np
import itertools

n = len(input_rays)
# n = 20
# n = 10
# n = 50
# n = 75

# Defining Z3 variables
x_vars = [Real(f'x_{i}') for i in range(n)]
y_vars = [Real(f'y_{i}') for i in range(n)]
z_vars = [Real(f'z_{i}') for i in range(n)]

v_x_hat = Int('v_x_hat')
v_y_hat = Int('v_y_hat')
v_z_hat = Int('v_z_hat')

# Z3 solver
solver = Solver()

# Add non-zero constraints
solver.add(Not(v_x_hat == 0))
solver.add(Not(v_y_hat == 0))
solver.add(Not(v_z_hat == 0))

# Translating SymPy equations to Z3
for i, (r, v) in enumerate(input_rays[:n]):
    solver.add((x_vars[i] - r[0])/v[0] == (y_vars[i] - r[1])/v[1])
    solver.add((x_vars[i] - r[0])/v[0] == (z_vars[i] - r[2])/v[2])
    solver.add((y_vars[i] - r[1])/v[1] == (z_vars[i] - r[2])/v[2])

# Generating all possible pairs
pairs = list(itertools.combinations(range(n), 2))
print(n, len(pairs))
# Randomly sample pairs
# sampled_pairs = np.random.choice(np.arange(len(pairs)), 500, replace=False)
sampled_pairs = np.random.choice(np.arange(len(pairs)), 10, replace=False)

# Pairwise comparisons for sampled pairs
for p in sampled_pairs:
# for p in range(len(pairs)):
    i, j = pairs[p]

    solver.add( ((y_vars[i] - y_vars[j]) * v_x_hat) == ((x_vars[i] - x_vars[j]) * v_y_hat) )
    solver.add( ((z_vars[i] - z_vars[j]) * v_x_hat) == ((x_vars[i] - x_vars[j]) * v_z_hat) )
    solver.add( ((z_vars[i] - z_vars[j]) * v_y_hat) == ((y_vars[i] - y_vars[j]) * v_z_hat) )

# Check if the system is solvable
if solver.check() == sat:
    model = solver.model()
    # return model
else:
    # return "No solution found"
    pass

model
```

```python
import itertools
from sympy import symbols, Eq, solve, Symbol

# n = len(input_rays)
n = 3

t_symbols = symbols(f't0:{n}', positive=True)

v_x_hat = Symbol('v_x_hat', integer=True)
v_y_hat = Symbol('v_y_hat', integer=True)
v_z_hat = Symbol('v_z_hat', integer=True)

r_x_hat = Symbol('r_x_hat', integer=True)
r_y_hat = Symbol('r_y_hat', integer=True)
r_z_hat = Symbol('r_z_hat', integer=True)

# Create a list to hold all equations
equations = []

for i, (r, v) in enumerate(input_rays[:n]):
    eq1 = Eq(r_x_hat + v_x_hat * t_symbols[i], r[0] + v[0] * t_symbols[i])
    eq2 = Eq(r_y_hat + v_y_hat * t_symbols[i], r[1] + v[1] * t_symbols[i])
    eq3 = Eq(r_z_hat + v_z_hat * t_symbols[i], r[2] + v[2] * t_symbols[i])

    equations.append(eq1)
    equations.append(eq2)
    equations.append(eq3)

solutions = solve(equations, t_symbols + (v_x_hat, v_y_hat, v_z_hat) + (r_x_hat, r_y_hat, r_z_hat))
solutions
```
