```python
def run_hash(s):
    v = 0

    for c in s:
        v += ord(c)
        v *= 17
        v %= 256
    
    return v
```

```python
boxes = [[] for _ in range(256)]

# going to do front, ... back order

# for s in test_input2.split(','):
for s in input.split(','):
    if '=' in s:
        label, num_str = s.split('=')
        focal_length = int(num_str)

        box_num = run_hash(label)

        for i, (existing_label, existing_focal_length) in enumerate(boxes[box_num]):
            if existing_label == label:
                boxes[box_num][i] = (label, focal_length)
                break
        else:
            boxes[box_num].append((label, focal_length))
    elif '-' in s:
        label = s[:-1]

        box_num = run_hash(label)

        boxes[box_num] = [(l, f) for l, f in boxes[box_num] if l != label]
```

```python
boxes
```

```python
sm = 0

for box_num, box in enumerate(boxes, start=1):
    for slot_num, (_, focal_length) in enumerate(box, start=1):
        sm += box_num * slot_num * focal_length

sm
```
