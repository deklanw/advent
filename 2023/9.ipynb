```python
test_input = """0 3 6 9 12 15
1 3 6 10 15 21
10 13 16 21 30 45"""
```

```python
import itertools
```

```python
input = open("inputs/9").read()
```

```python
def get_differences(lst):
    return [y - x for x, y in zip(lst, lst[1:])]
```

```python
histories = list(map(lambda x: list(map(int, x.split())), input.splitlines()))
```

```python
def p1(input):
    histories = list(map(lambda x: list(map(int, x.split())), input.splitlines()))

    s = 0

    for history in histories:
        higher_differences = [history]

        for _ in itertools.count():
            next_differences = get_differences(higher_differences[-1])
            higher_differences.append(next_differences)

            if all(d == 0 for d in next_differences):
                break
        
        higher_differences[-1].append(0)

        for higher_diff, lower_diff in itertools.pairwise(higher_differences[::-1]):
            lower_diff.append(lower_diff[-1] + higher_diff[-1])
        
        s += higher_differences[0][-1]
    
    return s
```

```python
p1(test_input), p1(input)
```

```python
def p2(input):
    histories = list(map(lambda x: list(map(int, x.split())), input.splitlines()))

    s = 0

    for history in histories:
        higher_differences = [history]

        for _ in itertools.count():
            next_differences = get_differences(higher_differences[-1])
            higher_differences.append(next_differences)

            if all(d == 0 for d in next_differences):
                break
        
        higher_differences[-1].append(0)

        # reverse each list
        for lst in higher_differences:
            lst.reverse()

        for higher_diff, lower_diff in itertools.pairwise(higher_differences[::-1]):
            # now we subtract!
            lower_diff.append(lower_diff[-1] - higher_diff[-1])
        
        s += higher_differences[0][-1]
    
    return s
```

```python
p2(test_input), p2(input)
```
