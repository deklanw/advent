```python
test_input = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#....."""

input = open("inputs/11").read()
```

```python
import numpy as np

def parse_board(input):
    return list(map(list, input.splitlines()))
```

```python
# board = np.array(parse_board(test_input))
board = np.array(parse_board(input))
board
```

```python
binary_board = (board == '#').astype(int)
binary_board
```

```python
column_sums = binary_board.sum(axis=0)
row_sums = binary_board.sum(axis=1)
```

```python
empty_rows = np.where(row_sums == 0)[0]
empty_cols = np.where(column_sums == 0)[0]
```

```python
empty_rows, empty_cols
```

```python
galaxy_row_idxs, galaxy_col_idxs = np.where(binary_board)
```

```python
# empty_replacement_factor = 2
# empty_replacement_factor = 10
# empty_replacement_factor = 100
empty_replacement_factor = 1000000
```

```python
row_offsets = (galaxy_row_idxs > empty_rows[:, None]).sum(axis=0) * (empty_replacement_factor - 1)
col_offsets = (galaxy_col_idxs > empty_cols[:, None]).sum(axis=0) * (empty_replacement_factor - 1)
```

```python
expanded_galaxy_row_idxs = row_offsets + galaxy_row_idxs
expanded_galaxy_col_idxs = col_offsets + galaxy_col_idxs
```

```python
import itertools


s = 0

for (p1, p2) in itertools.combinations(zip(expanded_galaxy_row_idxs, expanded_galaxy_col_idxs), 2):
    d = abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])
    s += d
```
