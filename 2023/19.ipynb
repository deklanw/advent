```python
input = open("inputs/19").read()
```

```python
def parse_ratings(rating_str):
    labels = ['x', 'm', 'a', 's']
    ints = [int(part.split('=')[1]) for part in rating_str.strip('{}').split(',')]

    return dict(zip(labels, ints))
```

```python
# workflows, ratings = test_input.split("\n\n")
workflows, ratings = input.split("\n\n")

workflows = workflows.split("\n")  
ratings = ratings.split("\n")
```

```python
workflow_dict = {}

for workflow in workflows:
    workflow_label, rest = workflow.split("{")
    rest = rest[:-1]

    rules = rest.split(",")

    if len(rules) > 3:
        # print(len(rules))
        print(rules)

    workflow_dict[workflow_label] = rules
```

```python
attr_sum = 0

for rating in ratings:
    rating = parse_ratings(rating)

    current_workflow = "in"
    while True:
        # print("current workflow", current_workflow)

        if current_workflow == "R":
            break

        if current_workflow == "A":
            attr_sum += sum(rating.values())
            break

        rules = workflow_dict[current_workflow]

        for rule in rules:
            if ':' in rule:
                condition, conditional_workflow = rule.split(':')

                lt = '<' in condition
                attribute, num = condition.split(lt and '<' or '>')
                num = int(num)

                if lt and rating[attribute] < num:
                    current_workflow = conditional_workflow
                    break
                elif not lt and rating[attribute] > num:
                    current_workflow = conditional_workflow
                    break
            else:
                current_workflow = rule

attr_sum
```

```python
MIN_VALUE = 1
MAX_VALUE = 4000

def ray_complement(interval):
    if interval[0] == MIN_VALUE and interval[1] == MAX_VALUE:
        assert False

    if interval[0] == MIN_VALUE:
        return (interval[1] + 1, MAX_VALUE)
    elif interval[1] == MAX_VALUE:
        return (MIN_VALUE, interval[0] - 1)
    else:
        assert False
```

```python
def oned_intersection(a, b):
    x1 = max(min(a), min(b))
    x2 = min(max(a), max(b))

    return (x1, x2) if x1 <= x2 else None
```

```python
complete_paths = []

def negate_condition(condition):
    attribute, interval = condition

    return attribute, ray_complement(interval)

def walk_tree(current_label, path_conditions):
    if current_label == "R":
        return
    
    if current_label == 'A':
        complete_paths.append(path_conditions)
        return

    rules = workflow_dict[current_label]

    edges = []
    conditions = []

    for rule in rules:
        negated_prior_conditions = [negate_condition(c) for c in conditions]

        if ':' in rule:
            condition, conditional_workflow = rule.split(':')

            lt = '<' in condition
            attribute, num = condition.split(lt and '<' or '>')
            num = int(num)

            interval = ( MIN_VALUE, num-1 ) if lt else ( num+1, MAX_VALUE )

            condition = (attribute, interval)

            edges.append((conditional_workflow, negated_prior_conditions + [condition]))
            conditions.append(condition)

        else:
            # no new conditions added, the whole rule is the next label
            edges.append((rule, negated_prior_conditions))

    for (next_label, transition_condition) in edges:
        walk_tree(next_label, path_conditions + transition_condition)
```

```python
import itertools
import functools

reduced_complete_paths = []

# now for each complete path let's group by attribute then do a 1d intersection of the intervals
for path in complete_paths:
    # God I hate python's groupby API. yes, the sorted is necessary
    grouped_by_attribute = itertools.groupby(sorted(path, key=lambda x: x[0]), lambda x: x[0])

    reduced_path = []

    for attribute, conditions in grouped_by_attribute:
        lst = [c[1] for c in conditions]
        reduced = functools.reduce(oned_intersection, lst)
        reduced_path.append((attribute, reduced))
        # print(lst)
        # print(reduced)
        # print()
    
    # print(path)
    # print(reduced_path)
    # print()

    reduced_complete_paths.append(reduced_path)
```

```python
sm = 0

letters = list('xmas')

for path in reduced_complete_paths:
    path_dict = dict(path)

    prod = 1
    for letter in letters:
        if letter in path_dict:
            interval = path_dict[letter]
        else:
            interval = (MIN_VALUE, MAX_VALUE)
    
        prod *= interval[1] - interval[0] + 1

    sm += prod
sm
```
