```python
import itertools

def solve_crm_sieve(equations):
    # takes pairs of (ai, n1)

    # first sort by magnitude of n1
    equations_sorted = sorted(equations, key=lambda x: x[1], reverse=True)
    print(equations_sorted)

    current_a, current_n = equations_sorted[0]
    # now first reduce everything such that ai < ni

    equations_sorted = [(a % n, n) for a, n in equations_sorted]

    for next_a, next_n in equations_sorted[1:]:
        for i in itertools.count():
            next_val = current_a + i*current_n

            if next_val % next_n == next_a:
                current_a = next_val
                current_n *= next_n
                break
    
    solution = next_val

    # now make sure it works
    for (a, n) in equations_sorted:
        assert solution % n == a

    return solution

solve_crm_sieve([(0, 3), (3, 4), (4, 5)])
```
