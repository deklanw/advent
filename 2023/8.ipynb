```python
test_input = """RL

AAA = (BBB, CCC)
BBB = (DDD, EEE)
CCC = (ZZZ, GGG)
DDD = (DDD, DDD)
EEE = (EEE, EEE)
GGG = (GGG, GGG)
ZZZ = (ZZZ, ZZZ)"""

test_input2 = """LLR

AAA = (BBB, BBB)
BBB = (AAA, ZZZ)
ZZZ = (ZZZ, ZZZ)"""
```

```python
input = open("inputs/8").read()
```

```python
def parse_map(input):
    sequence, map = input.split("\n\n")

    map_d = {}

    for ln in map.splitlines():
        start, tpl = ln.split(" = ")
        left_label, right_label = tpl[1:-1].split(", ")

        map_d[start] = (left_label, right_label)
    
    return sequence, map_d
```

```python
from itertools import cycle

def p1(input):
    sequence, map_d = parse_map(input)

    current_label = 'AAA'

    for i, dir in enumerate(cycle(sequence)):
        if current_label == 'ZZZ':
            break

        idx = int(dir == 'R')
        current_label = map_d[current_label][idx]
    
    return i
```

```python
p1(test_input), p1(input)
```

```python
test_input3 = """LR

11A = (11B, XXX)
11B = (XXX, 11Z)
11Z = (11B, XXX)
22A = (22B, XXX)
22B = (22C, 22C)
22C = (22Z, 22Z)
22Z = (22B, 22B)
XXX = (XXX, XXX)"""
```

```python
import math

def p2(input):
    sequence, map_d = parse_map(input)

    current_labels = [l for l in map_d.keys() if l[-1] == 'A']
    memories = [[] for _ in range(len(current_labels))]

    for i, dir in enumerate(cycle(sequence), start=1):
        idx = int(dir == 'R')

        current_labels = [map_d[l][idx] for l in current_labels]

        for j, l in enumerate(current_labels):
            if l[-1] == 'Z':
                memories[j].append(i)

        if all(len(lst) > 3 for lst in memories):
            print(f"Memories look ready, at iteration {i}")
            break

    ds = []
    for lst in memories:
        diffs = [y - x for x, y in zip(lst, lst[1:])]
        assert len(set(diffs)) == 1
        ds.append(diffs[0])
    
    return math.lcm(*ds)
```

```python
p2(test_input3), p2(input)
```
