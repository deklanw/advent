```python
test_input = """1abc2
pqr3stu8vwx
a1b2c3d4e5f
treb7uchet"""
```

```python
def pt1(input):
    s = 0

    for l in input.splitlines():
        nums = [s for s in l if s.isdigit()]
        final_num = int(nums[0] + nums[-1])
        s += final_num
    
    return s
```

```python
digit_strs = {s: str(i+1) for i, s in enumerate(["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"])}
```

```python
digit_strs
```

```python
def get_seen_digits(s):
    seen_digits = []

    for i, c in enumerate(s):
        if c.isdigit():
            seen_digits.append(str(c))
        else:
            for d_word, d_str in digit_strs.items():
                if len(d_word) + i <= len(s) and s[i:i+len(d_word)] == d_word:
                    seen_digits.append(d_str)
    return seen_digits
```

```python
def pt2(input):
    s = 0

    for l in input.splitlines():
        seen_digits = get_seen_digits(l)
        s += int(seen_digits[0] + seen_digits[-1])
    
    return s
```

```python
pt2(test_input2)
```
