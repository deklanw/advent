```python
# distance when holding for t seconds when race time is R for t in range [0, R], integers only
# t * (R-t)
# d = -t^2 + Rt
# so, when t=0, d=0. when t=1, d=R-1, when t=2 d=-4+2R. when t=R d=0.

# find critical point
# -2t + R = 0
# t = R/2

# given a distance let's find t with the quadratic equation
# d = -t^2 + Rt
# 0 = -t^2 + Rt - d

# t = (-R +- sqrt(R^2 - 4*d)) / -2

# t = (R +- sqrt(R^2 - 4*d)) / 2
```

```python
test_input = """Time:      7  15   30
Distance:  9  40  200"""
```

```python
input = open("inputs/6").read()
```

```python
def p1(input):
    times_line, distances_line = input.splitlines()

    times = list(map(int, times_line.split()[1:]))
    distances = list(map(int, distances_line.split()[1:]))
    races = list(zip(times, distances))

    prod = 1

    for (total_time, distance_record) in races:
        f = lambda t: - (t**2) + total_time*t

        num = 0
        for t in range(0, total_time+1):
            if f(t) > distance_record:
                num += 1
        
        prod *= num
    
    return prod
```

```python
p1(test_input), p1(input)
```

```python
def p2(input):
    times_line, distances_line = input.splitlines()

    total_time = int("".join(times_line.split()[1:]))
    distance_record = int("".join(distances_line.split()[1:]))

    f = lambda t: - (t**2) + total_time*t

    num = 0
    for t in range(0, total_time+1):
        if f(t) > distance_record:
            num += 1
    
    return num
```

```python
import math

def count_improvements(total_time, distance_record):
    f = lambda t: - (t**2) + total_time*t

    # t = (R +- sqrt(R^2 - 4*d)) / 2

    roots = []

    for sign in [-1, 1]:
        roots.append((total_time + sign* (total_time**2 - 4*distance_record)**0.5) / 2)

    # critical_point = distance_record / 2
    # roots, critical_point, f(roots[0]), f(roots[1]), f(critical_point)

    # from symmetry around critical point
    r1, r2 = math.ceil(roots[0]), math.floor(roots[1])

    return r2 - r1 + 1
```

```python
count_improvements(7, 9), count_improvements(15, 40)
```

```python
def p1_better(input):
    times_line, distances_line = input.splitlines()

    times = list(map(int, times_line.split()[1:]))
    distances = list(map(int, distances_line.split()[1:]))
    races = list(zip(times, distances))

    prod = 1

    for (total_time, distance_record) in races:
        prod *= count_improvements(total_time, distance_record)
    
    return prod
```

```python
p1_better(test_input), p1_better(input)
```

```python
def p2_better(input):
    times_line, distances_line = input.splitlines()

    total_time = int("".join(times_line.split()[1:]))
    distance_record = int("".join(distances_line.split()[1:]))

    return count_improvements(total_time, distance_record)
```
