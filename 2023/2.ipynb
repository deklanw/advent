```python
colors = ['red', 'green', 'blue']
```

```python
test_input = '''Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green
Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue
Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red
Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red
Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green'''
```

```python
import re
from collections import defaultdict

def parse_input(input):
    games = []

    for line in input.splitlines():
        game = []
        for round in line.split(';'):
            colors_seen = defaultdict(int)

            for color in colors:
                search_re = r'(\d+) ' + color

                match = re.search(search_re, round)

                if match:
                    colors_seen[color] = int(match.group(1))

            game.append(colors_seen)

        games.append(game)
    
    return games
```

```python
games = parse_input(test_input)
```

```python
def check_possibility(game, color_starts):
    for round in game:
        for color in colors:
            if round[color] > color_starts[color]:
                return False

    return True

def pt1(input, color_starts):
    games = parse_input(input)

    return sum(i+1 if check_possibility(game, color_starts) else 0 for i, game in enumerate(games))
```

```python
color_starts = {'red': 12, 'green': 13, 'blue': 14}
pt1(test_input, color_starts)
```

```python
color_starts = {'red': 12, 'green': 13, 'blue': 14}
pt1(input, color_starts)
```

```python
from math import prod

def pt2(input):
    games = parse_input(input)

    # ok we just need to take the maximum seen colors of each cube over the round
    s = 0
    for game in games:
        max_colors = defaultdict(int)

        for round in game:
            for color in colors:
                max_colors[color] = max(max_colors[color], round[color])
        
        s += prod(max_colors.values())
    
    return s
```
