```python
directions = {
    "right": (0, 1),
    "left": (0, -1),
    "up": (-1, 0),
    "down": (1, 0)
}

# from collections import deque
# directions_rotate = deque(['right', 'down', 'left', 'up'])
```

```python
direction_mirror_to_new_direction = dict()

direction_mirror_to_new_direction["right", "/"] = "up"
direction_mirror_to_new_direction["down", "/"] = "left"
direction_mirror_to_new_direction["left", "/"] = "down"
direction_mirror_to_new_direction["up", "/"] = "right"

direction_mirror_to_new_direction["right", "\\"] = "down"
direction_mirror_to_new_direction["down", "\\"] = "right"
direction_mirror_to_new_direction["left", "\\"] = "up"
direction_mirror_to_new_direction["up", "\\"] = "left"
```

```python
def add_tuples(tuple1, tuple2):
    return tuple(map(lambda x, y: x + y, tuple1, tuple2))
```

```python
def in_bounds(shape, pos):
    return all(0 <= pos[i] < shape[i] for i in range(len(shape)))
```

```python
def get_light_energy(board, start_pos, start_dir):
    visited_with_dir = set()

    def visit_position(pos, direction):
        nonlocal visited_with_dir

        if not in_bounds(board.shape, pos):
            return
        
        if (pos, direction) in visited_with_dir:
            # print("Already visited", pos, direction)
            return
        
        visited_with_dir.add((pos, direction))
        
        char = board[pos]

        if char == "\\" or char == "/":
            new_direction = direction_mirror_to_new_direction[direction, char]
            new_pos = add_tuples(pos, directions[new_direction])

            return visit_position(new_pos, new_direction)

        if char == '.':
            new_pos = add_tuples(pos, directions[direction])
            return visit_position(new_pos, direction)
        
        if char == '|':
            if direction == "up" or direction == "down":
                new_pos = add_tuples(pos, directions[direction])
                return visit_position(new_pos, direction)
            else:
                up_pos = add_tuples(pos, directions["up"])
                down_pos = add_tuples(pos, directions["down"])

                return visit_position(up_pos, "up") or visit_position(down_pos, "down")
        
        if char == '-':
            if direction == "left" or direction == "right":
                new_pos = add_tuples(pos, directions[direction])
                return visit_position(new_pos, direction)
            else:
                left_pos = add_tuples(pos, directions["left"])
                right_pos = add_tuples(pos, directions["right"])

                return visit_position(left_pos, "left") or visit_position(right_pos, "right")

    visit_position(start_pos, start_dir)

    return len(set(p for p, dir in visited_with_dir))
```

```python
max_energy = -1

for i in range(board.shape[0]):
    l = get_light_energy(board, (i, 0), "right")

    if l > max_energy:
        max_energy = l
    
    l = get_light_energy(board, (i, board.shape[1] - 1), "left")
    
    if l > max_energy:
        max_energy = l

for j in range(board.shape[1]):
    l = get_light_energy(board, (0, j), "down")

    if l > max_energy:
        max_energy = l
    
    l = get_light_energy(board, (board.shape[0] - 1, j), "up")
    
    if l > max_energy:
        max_energy = l
```
