```python
def oned_intersection(interval1, interval2):
    # boundaries inclusive
    # just checks if two intervals intersect, returns bool
    if interval1[0] > interval2[0]:
        interval1, interval2 = interval2, interval1
    
    return interval1[1] >= interval2[0]

def spans_intersection(spans1, spans2):
    return all((oned_intersection(s1, s2) for s1, s2 in zip(spans1, spans2)))
```

```python
# anything already at z=1 is at rest, otherwise we take the brick at the lowest z coord, figure out where it will rest. repeat until everything is at rest.
```

```python
import itertools
from collections import defaultdict
xy_intersections = defaultdict(list)

for (i, j) in itertools.combinations(range(len(bricks)), 2):
    if spans_intersection(brick_spans[i][:2], brick_spans[j][:2]):
        xy_intersections[i].append(j)
        xy_intersections[j].append(i)
```

```python
for i, current_brick in enumerate(bricks):

    brick_spans = [tuple(zip(*bs)) for bs in bricks]
    z_maxs = [b[-1][-1] for b in bricks]
    z_mins = [b[0][-1] for b in bricks]

    print(f"Brick {i}. Current zmax: {z_maxs[i]}. Current zmin: {z_mins[i]}")

    if z_mins[i] == 1:
        print("Already at rest")
        print()
        continue

    # otherwise let's find the highest brick under us which intersects in the x, y plane
    z_max_under, brick_under = max(((z_maxs[j], j) for j in xy_intersections[i] if z_maxs[j] < z_mins[i]), default=(0, None))

    if brick_under is not None:
        print(f"Found brick under at index {brick_under}")
        print(f"Their z max is {z_max_under}")
        print(f"Their spans: {brick_spans[brick_under]}. Our spans: {brick_spans[i]}")
        new_z = z_max_under + 1
    else:
        print("Nothing under. Hitting floor.")
        new_z = 1
    
    # now we modify the brick position: we want the bottom z coord to be new_z
    # so we add new_z - old_bottom_z to both coords
    to_add_z = new_z - current_brick[0][-1]

    print(f"Shifting brick {i} down by {to_add_z}")

    current_brick[0][-1] += to_add_z
    current_brick[1][-1] += to_add_z

    # now the loop continues until we've processed every brick

    print()
```

```python
from collections import defaultdict

z_min_to_idxs = defaultdict(list)
z_max_to_idxs = defaultdict(list)

for i, z in enumerate(z_mins):
    z_min_to_idxs[z].append(i)

for i, z in enumerate(z_maxs):
    z_max_to_idxs[z].append(i)
```

```python
g = nx.DiGraph()

# edge from a->b means a supports b
```

```python
import itertools

for z_max, max_idxs in z_max_to_idxs.items():
    above = z_max + 1

    if above not in z_min_to_idxs:
        continue

    min_idxs = z_min_to_idxs[above]

    for (max_idx, min_idx) in itertools.product(max_idxs, min_idxs):
        if spans_intersection(brick_spans[max_idx][:2], brick_spans[min_idx][:2]):
            # g.add_edge(max_idx, min_idx)
            g.add_edge(min_idx, max_idx)
```

```python
# sum of all bricks whose in-neighbors all have more than out-degree 1

# 397 too low
# 450 is too high

def count_non_essential_nodes(g):
    count = 0
    for node in g.nodes():
        in_neighbors = g.predecessors(node)
        if all(g.out_degree(neighbor) > 1 for neighbor in in_neighbors):
            count += 1
    return count

count_non_essential_nodes(g) + isolated_bricks
```
