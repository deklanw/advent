```python
from collections import Counter
```

```python
instructions = [s.split() for s in open('./inputs/6', mode="r").read().split('\n\n')]
```

```python
instructions
```

```python
count = 0

for group in instructions:
    yes_questions = set()
    
    for person_yes in group:
        # set constructor automatically separates string into characters
        yes_questions |= set(person_yes)

    count += len(yes_questions)
```

```python
count
```

```python
count = 0

for group in instructions:
    # initialize with anyone from group
    yes_questions = set(group[0])
    
    for person_yes in group:
        # set constructor automatically separates string into characters
        yes_questions &= set(person_yes)

    count += len(yes_questions)
```
