```python
def follow_instructions(instructions):
    # going by unit circle convention in degrees
    # this is east
    angle = 0
    position = complex(0,0)

    for (inst, n) in instructions:
        if inst == 'N':
            position += complex(0, n)
        elif inst == 'S':
            position -= complex(0, n)
        elif inst == 'E':
            position += complex(n, 0)
        elif inst == 'W':
            position -= complex(n, 0)
        elif inst == 'L':
            angle += n
        elif inst == 'R':
            angle -= n
        elif inst == 'F':
            rad = math.radians(angle)
            position += cmath.rect(n, rad)
        else:
            raise Exception("Unknown command", inst)

    return np.abs(position.real) + np.abs(position.imag)
```

```python
def follow_instructions2(instructions):        
    ship_position = complex(0., 0.)
    way_point_rel_position = complex(10., 1.)
    
    for (inst, n) in instructions:
        if inst == 'N':
            way_point_rel_position += complex(0, n)
        elif inst == 'S':
            way_point_rel_position -= complex(0, n)
        elif inst == 'E':
            way_point_rel_position += complex(n, 0)
        elif inst == 'W':
            way_point_rel_position -= complex(n, 0)
        elif inst == 'L' or inst == 'R':
            delta_angle = n if inst == 'L' else -n
            delta_rad = math.radians(delta_angle)
            r, current_phi = cmath.polar(way_point_rel_position)
            way_point_rel_position = cmath.rect(r, current_phi + delta_rad)
        elif inst == 'F':
            ship_position += n * way_point_rel_position
        else:
            raise Exception("Unknown command", inst)
        
    
    return np.abs(ship_position.real) + np.abs(ship_position.imag)
```
