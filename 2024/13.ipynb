```python
import re


def parse_input(input_str):
    data_blocks = input_str.strip().split("\n\n")
    result = []

    for block in data_blocks:
        lines = block.split("\n")
        button_a_coords = tuple(
            map(int, re.findall(r"X\+(\d+), Y\+(\d+)", lines[0])[0])
        )
        button_b_coords = tuple(
            map(int, re.findall(r"X\+(\d+), Y\+(\d+)", lines[1])[0])
        )
        prize_coords = tuple(map(int, re.findall(r"X=(\d+), Y=(\d+)", lines[2])[0]))
        result.append((button_a_coords, button_b_coords, prize_coords))

    return result


# Usage:
parsed_input = parse_input(input)
parsed_input
```

```python
def solve_lines(line1, line2):
    """solve using cramer's rule"""
    a1, b1, c1 = line1
    a2, b2, c2 = line2

    D = a1 * b2 - a2 * b1
    if D == 0:
        # Lines are parallel, check if they're the same line
        if a1 != 0:  # Avoid division by zero
            ratio = a2 / a1
            if abs(b2 / b1 - ratio) < 1e-10 and abs(c2 / c1 - ratio) < 1e-10:
                print("infinite intersection")
                return "infinite"
        print("no intersection")
        return "none"

    x = (b1 * c2 - b2 * c1) / D
    y = (c1 * a2 - c2 * a1) / D
    return x, y


total_cost = 0

for line in parsed_input:
    (a1, b1), (a2, b2), (goal_x, goal_y) = line

    unique_solution = solve_lines((a1, a2, -goal_x), (b1, b2, -goal_y))

    if not unique_solution:
        continue

    if all(r.is_integer() for r in unique_solution):
        total_cost += 3 * int(unique_solution[0]) + int(unique_solution[1])

total_cost
```

```python
FUDGE_FACTOR = 10000000000000
total_cost = 0

for line in parsed_input:
    (a1, b1), (a2, b2), (goal_x, goal_y) = line

    unique_solution = solve_lines(
        (a1, a2, -(goal_x + FUDGE_FACTOR)), (b1, b2, -(goal_y + FUDGE_FACTOR))
    )

    if not unique_solution:
        continue

    if all(r.is_integer() for r in unique_solution):
        total_cost += 3 * int(unique_solution[0]) + int(unique_solution[1])

total_cost
```
