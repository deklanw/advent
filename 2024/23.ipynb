```python
connections = set(map(tuple, [s.split("-") for s in input.split()]))
connections
```

```python
all_nodes = set()

for c1, c2 in connections:
    all_nodes.add(c1)
    all_nodes.add(c2)

len(all_nodes)
```

```python
import itertools

triangles = set()


def check_edge(a, b):
    return (a, b) in connections or (b, a) in connections


for v, (c1, c2) in itertools.product(all_nodes, connections):
    if check_edge(v, c1) and check_edge(v, c2):
        triangles.add(tuple(sorted([v, c1, c2])))
```

```python
import itertools


def check_edge(a, b):
    return (a, b) in connections or (b, a) in connections


current_cliques = connections

for clique_size in itertools.count(3):
    new_cliques = set()

    for v, cs in itertools.product(all_nodes, current_cliques):
        if all(check_edge(v, c) for c in cs):
            new_cliques.add(tuple(sorted([v, *cs])))

    if len(new_cliques) == 0:
        print(f"No cliques of size {clique_size}, exiting")
        break
    else:
        print(f"{len(new_cliques)} cliques of size {clique_size}, continuing")

    current_cliques = new_cliques
```
