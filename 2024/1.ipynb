```python
from collections import Counter
```

```python
pairs = [tuple(map(int, line.split())) for line in input.splitlines()]
pairs[-3:]
```

```python
lst1 = [p[0] for p in pairs]
lst2 = [p[1] for p in pairs]

lst1[:2], lst2[:2]
```

```python
sum(abs(a - b) for a, b in zip(sorted(lst1), sorted(lst2)))
```

```python
lst2_counter = Counter(lst2)
lst2_counter.most_common()[:5]
```

```python
s = 0

for n in lst1:
    s += n * lst2_counter[n]

s
```
