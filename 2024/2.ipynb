```python
input = open("inputs/2", "r").read()
```

```python
import numpy as np


def parse_board(input):
    return list(map(lambda s: ([int(s) for s in s.split()]), input.splitlines()))


# boards = parse_board(test_input)
boards = parse_board(input)


def is_valid(lst):
    lst = np.array(lst)
    diff = lst[:-1] - lst[1:]
    return (np.abs(diff) <= 3).all() and ((diff > 0).all() or (diff < 0).all())


sum(is_valid(lst) for lst in boards)
```

```python
s = 0

for lst in boards:
    # when i=0 we're checking the original list
    for i in range(len(lst)):
        new_arr = lst[:i] + lst[i + 1 :]

        if is_valid(new_arr):
            s += 1
            break

s
```
