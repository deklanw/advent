```python
input = open("inputs/25").read()
```

```python
import numpy as np


def parse_board(input):
    return np.array(list(map(list, input.splitlines())))
```

```python
boards = [parse_board(s) for s in input.split("\n\n")]
```

```python
boards[0]
```

```python
locks = np.stack([b for b in boards if b[0, 0] == "#"], axis=-1)
keys = np.stack([b for b in boards if b[0, 0] == "."], axis=-1)
```

```python
locks.shape, keys.shape
```

```python
lock_cols = (locks == "#").sum(axis=0)
key_cols = (keys == "#").sum(axis=0)
lock_cols.shape, key_cols.shape
```

```python
fits = ((lock_cols[:, :, None] + key_cols[:, None, :]) <= locks.shape[0]).all(axis=0)
fits.shape
```
