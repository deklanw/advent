```python
import numpy as np
from itertools import combinations


def parse_board(input):
    return np.array(list(map(list, input.splitlines())))


def where_to_tuples(mask):
    return [tuple(map(int, t)) for t in zip(*np.where(mask))]


def add_tuple(t1, t2):
    return tuple(x + y for x, y in zip(t1, t2))


def neg_tuple(t):
    return tuple(-x for x in t)


def in_bounds(location, bounds):
    return all(0 <= x < y for x, y in zip(location, bounds))


def generate_discrete_line(start, dir, shape):
    location = start
    yield location

    for dir in [dir, neg_tuple(dir)]:
        while in_bounds(add_tuple(location, dir), shape):
            location = add_tuple(location, dir)
            yield location
```

```python
board = parse_board(input)
```

```python
ant_locations = where_to_tuples(board != ".")
```

```python
from collections import defaultdict

ant_locations_d = defaultdict(list)

for i, j in ant_locations:
    v = board[i, j]
    ant_locations_d[v].append((i, j))
```

```python
an_locations = []

for letter, letter_antennas in ant_locations_d.items():
    for a, b in list(combinations(letter_antennas, 2)):
        v_ba = add_tuple(b, neg_tuple(a))

        an_one = add_tuple(a, neg_tuple(v_ba))
        an_two = add_tuple(b, v_ba)

        if in_bounds(an_one, board.shape):
            an_locations.append([an_one, letter])

        if in_bounds(an_two, board.shape):
            an_locations.append([an_two, letter])

len(set(p for (p, _l) in an_locations))
```

```python
an_locations = []

for letter, letter_antennas in ant_locations_d.items():
    for a, b in list(combinations(letter_antennas, 2)):
        v_ba = add_tuple(b, neg_tuple(a))

        for ant_location in generate_discrete_line(a, v_ba, board.shape):
            an_locations.append([ant_location, letter])

len(set(p for (p, _l) in an_locations))
```
