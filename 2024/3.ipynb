```python
import re
```

```python
input = open("inputs/3").read()
```

```python
# input = "xmul(2,4)%&mul[3,7]!@^do_not_mul(5,5)+mul(32,64]then(mul(11,8)mul(8,5))"
# input = "xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))"
```

```python
pattern = r"mul\((\d{1,3}),(\d{1,3})\)"
```

```python
s = 0
for a, b in [map(int, m) for m in re.findall(pattern, input)]:
    s += a * b
s
```

```python
pattern2 = r"(do\(\))|(don\'t\(\))|mul\((\d{1,3}),(\d{1,3})\)"
```

```python
s = 0
enabled = True

for do, dont, a, b in re.findall(pattern2, input):
    if do:
        enabled = True
    elif dont:
        enabled = False
    elif enabled:
        s += int(a) * int(b)
s
```
