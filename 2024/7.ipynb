```python
lines = input.split("\n")

problems = []
for line in lines:
    total, rest = line.split(": ")
    total = int(total)
    nums = list(map(int, rest.split(" ")))

    problems.append((total, nums))
```

```python
def try_operation(goal, numbers):
    return operation_recurse(numbers[0], goal, numbers[1:])


def operation_recurse(current, goal, remaining):
    if len(remaining) == 0:
        return current == goal

    # optimization
    if current > goal:
        return False

    next = remaining[0]

    mul = operation_recurse(current * next, goal, remaining[1:])
    add = operation_recurse(current + next, goal, remaining[1:])

    return mul or add
```

```python
s = 0
for total, nums in problems:
    if try_operation(total, nums):
        s += total
s
```

```python
def try_operation_pt2(goal, numbers):
    return operation_recurse_pt2(numbers[0], goal, numbers[1:])


def operation_recurse_pt2(current, goal, remaining):
    if len(remaining) == 0:
        return current == goal

    # optimization
    if current > goal:
        return False

    next = remaining[0]

    mul = operation_recurse_pt2(current * next, goal, remaining[1:])
    add = operation_recurse_pt2(current + next, goal, remaining[1:])
    cat = operation_recurse_pt2(int(str(current) + str(next)), goal, remaining[1:])

    return mul or add or cat
```

```python
s = 0
for total, nums in problems:
    if try_operation_pt2(total, nums):
        s += total
s
```
