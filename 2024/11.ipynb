```python
stones = [int(s) for s in input.split()]
stones
```

```python
from functools import lru_cache

MULT = 2024


@lru_cache(maxsize=100_000)
def count_stones(stone, remaining_steps):
    if remaining_steps == 0:
        return 1

    if stone == 0:
        return count_stones(1, remaining_steps - 1)

    stone_str = str(stone)

    if len(stone_str) % 2 == 0:
        midpoint = len(stone_str) // 2
        stone_a = int(stone_str[:midpoint])
        stone_b = int(stone_str[midpoint:])

        return count_stones(stone_a, remaining_steps - 1) + count_stones(
            stone_b, remaining_steps - 1
        )
    else:
        return count_stones(stone * MULT, remaining_steps - 1)
```

```python
sum(count_stones(stone, 25) for stone in stones)
```
